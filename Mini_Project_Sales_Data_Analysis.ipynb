**Sales Data Analysis**  
   - Create a DataFrame of sales (columns: Product, Price, Quantity).  
   - Calculate total sales for each product.  
   - Plot a bar chart of sales using Matplotlib.  


```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Printer'],
    'Price': [1000, 25, 45, 150, 120],
    'Quantity': [10, 150, 85, 30, 20]
}

df = pd.DataFrame(data)

df['Total Sales'] = df['Price'] * df['Quantity']

print(df)


plt.bar(df['Product'], df['Total Sales'], color='skyblue')
plt.xlabel('Product')
plt.ylabel('Total Sales ($)')
plt.title('Total Sales by Product')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
