**Student Performance Dashboard**  
   - Create DataFrame of students with scores in 3 subjects.  
   - Compute averages and assign grades.  
   - Use Seaborn to visualize score distribution.  

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Math': [92, 78, 85, 60, 99],
    'Science': [88, 80, 79, 70, 95],
    'English': [90, 75, 82, 65, 91]
}

df = pd.DataFrame(data)


df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)


def assign_grade(avg):
    if avg >= 90:
        return 'A'
    elif avg >= 80:
        return 'B'
    elif avg >= 70:
        return 'C'
    elif avg >= 60:
        return 'D'
    else:
        return 'F'

df['Grade'] = df['Average'].apply(assign_grade)

print(df)


plt.figure(figsize=(12, 5))


plt.subplot(1, 2, 1)
sns.histplot(df[['Math', 'Science', 'English']], kde=True, bins=10)
plt.title("Score Distribution Across Subjects")

plt.subplot(1, 2, 2)
sns.boxplot(data=df[['Math', 'Science', 'English']])
plt.title("Boxplot of Scores by Subject")

plt.tight_layout()
plt.show()
```
